## Objective
Analyze retail sales data to identify top performing categories, 
regions, and understand the impact of discounting on profit.

## Key Questions
1. Which category generates the most sales?
2. Which region is most profitable?
3. Does discounting hurt profit?
4. What are the top sub-categories by revenue?
5. How is sales distributed across customer segments?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/SampleSuperstore.csv',encoding='latin-1')

# Verify it works
print(df.shape)
print(df.columns.tolist())
```

```python
# Chart 1 - Sales by Category
cat_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
sns.barplot(x=cat_sales.index, y=cat_sales.values, palette='Blues_d')
plt.title('Sales by Category')
plt.savefig('charts/chart1_category.png')
plt.show()
```

```python
# Chart 2 - Profit by Region
region_profit = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
sns.barplot(x=region_profit.index, y=region_profit.values, palette='Greens_d')
plt.title('Profit by Region')
plt.savefig('charts/chart2_region.png')
plt.show()
```

```python
# Chart 3 - Discount vs Profit
sns.scatterplot(data=df, x='Discount', y='Profit', alpha=0.5, color='coral')
plt.title('Discount vs Profit')
plt.savefig('charts/chart3_discount.png')
plt.show()
```

```python
# Chart 4 - Top Sub-Categories
sub_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=cat_sales.index, y=cat_sales.values, hue=cat_sales.index, palette='Blues_d', legend=False)
plt.title('Top 10 Sub-Categories by Sales')
plt.savefig('charts/chart4_subcategory.png')
plt.show()
```

```python
# Chart 5 - Segment Pie
seg_sales = df.groupby('Segment')['Sales'].sum()
plt.pie(seg_sales, labels=seg_sales.index, autopct='%1.1f%%')
plt.title('Sales by Segment')
plt.savefig('charts/chart5_segment.png')
plt.show()
```

## Key Findings & Insights

1. **Technology** is the highest selling category overall
2. **West region** generates the most profit among all regions
3. **High discounts (above 0.4) consistently result in losses** — 
   the scatter plot clearly shows profit goes negative as discount increases
4. **Phones and Chairs** are the top sub-categories by revenue
5. **Consumer segment** accounts for the largest share of total sales

## Business Recommendation

Furniture category where discounts are causing profit loss.
Focus more resources on Technology products in the West region 
for maximum returns.
